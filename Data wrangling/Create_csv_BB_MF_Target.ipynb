```python


def load_bb_and_add_to_df(mfqFileFullPath,df_ff):

    print('add_to_csv|| mfqFileFullPath: ',mfqFileFullPath)

    filename_w_ext = os.path.basename(mfqFileFullPath)
    filename, file_extension = os.path.splitext(filename_w_ext)

    indice = 'cv-valid-train/' + filename + '.mp3'
    
    print('indice:',indice)
    
    my_int = np.loadtxt(mfqFileFullPath,dtype=np.float64)

    print('my_int: ',my_int)
    
    df_ff.loc[indice,'pow'] = my_int[0]
    df_ff.loc[indice,'mfq'] = my_int[1]
    df_ff.loc[indice,'std'] = my_int[2]
    
    return df_ff

    
```

```python
for k,file in enumerate(os.listdir(sound_dir)):
    filename = os.fsdecode(file)
    full_path = os.path.join(sound_dir, file)

    filename_w_ext = os.path.basename(full_path)
    filename, file_extension = os.path.splitext(filename_w_ext)
    
    if file_extension == '.mfq':    
        df_full =  load_bb_and_add_to_df(full_path,df_full)
        
```
